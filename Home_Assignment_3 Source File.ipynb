Q1.  **Implementing a Basic Autoencoder**

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.datasets import mnist

# Load MNIST dataset
(x_train, _), (x_test, _) = mnist.load_data()

# Normalize and flatten the images
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

# Function to build an autoencoder
def build_autoencoder(latent_dim):
    # Encoder
    input_img = Input(shape=(784,))
    encoded = Dense(latent_dim, activation='relu')(input_img)

    # Decoder
    decoded = Dense(784, activation='sigmoid')(encoded)

    # Autoencoder model
    autoencoder = Model(input_img, decoded)

    # Compile model
    autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

    return autoencoder

# Function to train and evaluate autoencoder
def train_and_evaluate(latent_dim):
    autoencoder = build_autoencoder(latent_dim)

    # Train autoencoder
    autoencoder.fit(x_train, x_train, epochs=10, batch_size=256, shuffle=True, validation_data=(x_test, x_test), verbose=1)

    # Get reconstructed images
    reconstructed = autoencoder.predict(x_test)

    # Plot original and reconstructed images
    n = 10
    plt.figure(figsize=(20, 4))
    for i in range(n):
        # Original images
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
        plt.axis('off')

        # Reconstructed images
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(reconstructed[i].reshape(28, 28), cmap='gray')
        plt.axis('off')

    plt.show()

# Experiment with different latent dimensions
for latent_dim in [16, 32, 64]:
    print(f"Training Autoencoder with latent dimension: {latent_dim}")
    train_and_evaluate(latent_dim)
```

Q2. **Implementing a Denoising Autoencoder**

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.datasets import mnist

# Load MNIST dataset
(x_train, _), (x_test, _) = mnist.load_data()

# Normalize and flatten the images
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

# Add Gaussian noise to the images
noise_factor = 0.5
x_train_noisy = x_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape)
x_test_noisy = x_test + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape)

# Clip values to stay within [0,1]
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)

# Function to build a denoising autoencoder
def build_denoising_autoencoder(latent_dim):
    # Encoder
    input_img = Input(shape=(784,))
    encoded = Dense(latent_dim, activation='relu')(input_img)

    # Decoder
    decoded = Dense(784, activation='sigmoid')(encoded)

    # Autoencoder model
    autoencoder = Model(input_img, decoded)

    # Compile model
    autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

    return autoencoder

# Function to train and evaluate denoising autoencoder
def train_and_evaluate_denoising(latent_dim):
    autoencoder = build_denoising_autoencoder(latent_dim)

    # Train autoencoder with noisy input and clean output
    autoencoder.fit(x_train_noisy, x_train, epochs=10, batch_size=256, shuffle=True, validation_data=(x_test_noisy, x_test), verbose=1)

    # Get reconstructed images
    reconstructed = autoencoder.predict(x_test_noisy)

    # Plot noisy and reconstructed images
    n = 10
    plt.figure(figsize=(20, 4))
    for i in range(n):
        # Noisy images
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(x_test_noisy[i].reshape(28, 28), cmap='gray')
        plt.axis('off')

        # Reconstructed images
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(reconstructed[i].reshape(28, 28), cmap='gray')
        plt.axis('off')

    plt.show()

# Experiment with different latent dimensions
for latent_dim in [16, 32, 64]:
    print(f"Training Denoising Autoencoder with latent dimension: {latent_dim}")
    train_and_evaluate_denoising(latent_dim)

# Real-world application example:
# Denoising autoencoders are useful in medical imaging to remove noise from MRI or X-ray scans,
# improving clarity and aiding in accurate diagnosis.
```

Q3. **Implementing an RNN for Text Generation**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Dense, Embedding, Input
from tensorflow.keras.models import Sequential
import random
import sys

# Load dataset (Shakespeare's text)
path = tf.keras.utils.get_file("shakespeare.txt", "https://storage.googleapis.com/download.tensorflow.org/data/shakespeare.txt")
with open(path, "r", encoding="utf-8") as f:
    text = f.read().lower()

# Create character mapping
chars = sorted(set(text))
char_to_idx = {c: i for i, c in enumerate(chars)}
idx_to_char = {i: c for i, c in enumerate(chars)}

# Convert text to integer sequences
seq_length = 40  # Length of input sequences
step = 3  # Step size for sampling sequences
sequences = []
next_chars = []

for i in range(0, len(text) - seq_length, step):
    sequences.append(text[i: i + seq_length])
    next_chars.append(text[i + seq_length])

# Convert sequences to numerical format
X = np.zeros((len(sequences), seq_length, len(chars)), dtype=np.bool_)
y = np.zeros((len(sequences), len(chars)), dtype=np.bool_)

for i, sequence in enumerate(sequences):
    for t, char in enumerate(sequence):
        X[i, t, char_to_idx[char]] = 1
    y[i, char_to_idx[next_chars[i]]] = 1

# Define LSTM model
model = Sequential([
    LSTM(128, input_shape=(seq_length, len(chars))),
    Dense(len(chars), activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam')

# Train the model
model.fit(X, y, batch_size=128, epochs=10)

# Function to generate text
def generate_text(seed_text, length=400, temperature=1.0):
    generated = seed_text
    for _ in range(length):
        x_pred = np.zeros((1, seq_length, len(chars)))
        for t, char in enumerate(seed_text):
            x_pred[0, t, char_to_idx[char]] = 1

        preds = model.predict(x_pred, verbose=0)[0]

        # Apply temperature scaling
        preds = np.asarray(preds).astype('float64')
        preds = np.log(preds + 1e-8) / temperature  # Avoid log(0)
        exp_preds = np.exp(preds)
        preds = exp_preds / np.sum(exp_preds)

        next_idx = np.random.choice(len(chars), p=preds)
        next_char = idx_to_char[next_idx]

        seed_text = seed_text[1:] + next_char
        generated += next_char

    return generated

# Generate new text
seed = "to be, or not to be, that is the quest"
print(generate_text(seed, length=400, temperature=0.5))  # Adjust temperature for randomness
```

Q4. **Sentiment Classification Using RNN**

```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the IMDB dataset
num_words = 10000  # Vocabulary size
max_length = 200  # Max review length

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=num_words)

# Step 2: Preprocess the data
x_train = pad_sequences(x_train, maxlen=max_length)
x_test = pad_sequences(x_test, maxlen=max_length)

# Step 3: Build the LSTM model
model = Sequential([
    Embedding(input_dim=num_words, output_dim=128, input_length=max_length),
    LSTM(128, dropout=0.2, recurrent_dropout=0.2),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=3, batch_size=64, validation_data=(x_test, y_test))

# Step 4: Evaluate the model
y_pred = (model.predict(x_test) > 0.5).astype("int32")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Classification report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Step 5: Precision-Recall Tradeoff
print("""
Precision-Recall Tradeoff:
- High Precision: Less false positives, but might miss some positive reviews.
- High Recall: Catches more positive reviews, but might have more false positives.
- Sentiment analysis should balance both depending on use case.
""")
```
